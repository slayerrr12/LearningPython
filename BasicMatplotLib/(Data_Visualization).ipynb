### 1.3 Plot Multiple Lines

```python

x = [1,2,3,4,5]
y = [50,40,70,80,20]
y2 = [80,20,20,50,60]
y3 = [70,20,60,40,60]
y4 = [80,20,20,50,60]
plt.plot(x,y4,'y',label='KTM',linewidth=10)
plt.plot(x,y,'g',label='Enfield', linewidth=5)

plt.plot(x,y2,'c',label='Honda',linewidth=5)
plt.plot(x,y3,'k',label='Yahama',linewidth=5)
#plt.plot(x,y4,'y',label='KTM',linewidth=5)
plt.title('bike details in line plot')
plt.ylabel('Distance in kms')
plt.xlabel('Days')
plt.legend()
#plt.grid()
#plt.savefig("multipleline.jpg")
```

# 2 Bar Plot

### 2.1 Plot a Bar Chart

### 2.2 Multiple Bar Plot

```python
# Example 1
plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],
label="Enfield",width=.5)
plt.bar([0.26,1.25,2.25,3.25,4.25],[80,20,20,50,60],
label="Honda", color='r',width=.5)
plt.show()
plt.bar([0.31,1.5,2.5,3.5,4.5],[70,20,60,40,60],
label="Yamaha", color='y',width=.5)
plt.bar([.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],
label="KTM", color='g',width=.5)
plt.legend()
plt.xlabel('Days')
plt.ylabel('Distance (kms)')
plt.title('Bikes details in BAR PLOTTING')
```

```python
plt.bar(['A','B','C','D','E'],[50,40,70,80,20],
label="Enfield",width=.5)
plt.bar(['F','G','H','I','J'],[80,20,20,50,60],
label="Honda", color='r',width=.5)
plt.bar(['K','L','M','N','O'],[70,20,60,40,60],
label="Yamaha", color='y',width=.5)
plt.bar(['P','Q','R','S','T'],[80,20,20,50,60],
label="KTM", color='g',width=.5)
plt.legend()
plt.xlabel('Days')
plt.ylabel('Distance (kms)')
plt.title('Bikes details in BAR PLOTTING')
```

```python
#Example 2
from matplotlib import pyplot as plt
plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20], label="BMW")
plt.bar([.75,1.75,2.75,3.75,4.75],[80,20,20,50,60], label="Audi")
plt.plot([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],label = "BMW",color = 'red')
plt.plot([.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],label = "Audi",color = 'green')
plt.legend()
plt.grid()
plt.xlabel('Days')
plt.ylabel('Distance (kms)')
plt.title('Information')
plt.show()
```

### 2.3 Horizontal Bar Plot

# 3 Scatter Plot

### 3.1 Scatter plot withhout Details

```python


x = [5, 2, 9, 4, 7]     # x-axis values 

y = [10, 5, 8, 4, 2]         # Y-axis values 

plt.scatter(x, y)  
plt.grid()
# Function to plot scatter 

plt.show()               # function to show the plot 
```

### 3.2 Multiple Scatter Plot

```python


x = [1,1.5,2,2.5,3,3.5,3.6]
y = [7.5,8,8.5,9,9.5,10,10.5]
x1=[8,8.5,9,9.5,10,10.5,11]
y1=[3,3.5,3.7,4,4.5,5,5.2]
plt.scatter(x,y, label='high income low saving',color='r')
plt.scatter(x1,y1,label='low income high savings',color='b')
plt.xlabel('saving*100')
plt.ylabel('income*1000')
plt.title('Scatter Plot')
plt.legend()
#plt.show()       
plt.grid()
```

# 4  pie plot

```python
import matplotlib.pyplot as plt

days = [1,2,3,4]

slices = [1,2,3,4]
activities = ['Enfield','Honda','Yahama','KTM']
cols = ['r','g','y','b']
labels=activities
colors=cols
#slices
plt.pie(slices, labels = labels , colors = cols , startangle=60, shadow= True,explode=(0,.2,0,0), autopct='%1.2f%%')

plt.title('Bike details in Pie Plot')
```

### 5 Matplotlib Plots on NBA DataSet

```python

import pandas as pd
#import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('nba.csv')
df
```

### 5.1 Bar Graph

```python
#Example 1
y_values = df['Salary']
x_values = df['Age']
plt.xlabel('Age')                                   #To plot a bar graph plt.bar() command is used
plt.ylabel('Salary (in millions)')
plt.bar(x_values,y_values,color = "purple")          #This plots a bar graph between Age and Salaries of NBA players
plt.show()
```

```python
# Example 2
y_values = df['Salary']
x_values = df['Age']
plt.xlabel('Age')                           #making changes in the color field changes the colour of the graph     
plt.ylabel('Salary (in millions)')                  
plt.bar(x_values,y_values,color = "khaki")          
plt.show()
```

### 5.2 Horizontal Bar Graph

```python

y_values = df['Salary']
x_values = df['Age']
plt.xlabel('Age')                                         
plt.ylabel('Salary (in millions)')
plt.barh(x_values,y_values,color = "indianred")           
plt.show()
```

### 5.3 Scatter Plot

```python
#Example 1
y_value=df['Salary']
x_values=df['Age']                  #To display a scatter plot we use plt.scatter() command 
plt.xlabel('Age')
plt.ylabel('Salary (in millions)')
plt.scatter(x_values,y_values)      #This displays a scatter plot between Age and Salaries of NBA players
```

```python
#Example 2

y_value=df['Salary']
x_values=df['Age']
plt.xlabel('Age')
plt.ylabel('Salary (in millions)')
plt.scatter(x_values,y_values, color = "slategrey", marker = "X") #We can also change the type and colors of scatter plot as shown
```
